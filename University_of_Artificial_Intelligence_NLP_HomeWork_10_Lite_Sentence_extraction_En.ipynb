```
from transformers import T5Tokenizer, TFT5ForConditionalGeneration

import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
# https://pandas.pydata.org/pandas-docs/version/0.23.2/io.html#io-read-csv-table
df_titles = pd.read_csv('/content/drive/My Drive/Базы/NLP/nature_titles.txt', sep='\t', encoding = "UTF-8", quoting=3, doublequote=False)
df_paragraphs = pd.read_csv('/content/drive/My Drive/Базы/NLP/nature_articles.txt', sep='\t', encoding = "UTF-8", quoting=3, doublequote=False)
```

```
df_titles
```

```
count_titles = len(df_titles)
count_titles
```

```
df_paragraphs
```

```
count_paragraphs = len(df_paragraphs)
count_paragraphs
```

```
for i in range(count_titles):
  df_paragraphs["t"+str(i)] = ""
```

```
df_paragraphs
```

```
tokenizer = T5Tokenizer.from_pretrained('t5-large') # t5-small
model = TFT5ForConditionalGeneration.from_pretrained('t5-large') # t5-small
```

```
for i,title in enumerate(df_titles['title']):
  print(title)
  for j,paragraph in enumerate(df_paragraphs['paragraph']):
    command = f'stsb sentence1: {title} sentence2: {paragraph}'
    print(paragraph)
    input_ids = tokenizer.encode(command, return_tensors='tf')
    outputs = model.generate(input_ids)
    # print(outputs)
    result = tokenizer.decode(outputs[0])
    # print(result)
    df_paragraphs["t"+str(i)][j] = result
    print('title:', i, 'paragraph:', j, 'result=', result)
```

```
a = []
print('Results for sentence similarity (> 3.0)')
for i,title in enumerate(df_titles['title']):
  print(i+1, title)
  for j,paragraph in enumerate(df_paragraphs['paragraph']):
    if float(df_paragraphs["t"+str(i)][j]) > 3.0:
      print(' '*8, j, paragraph)
      a.append(j)
  print('-'*25)
```
